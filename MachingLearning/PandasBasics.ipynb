```python
import pandas as pd

personas = {
    "nombre" : ["Dimas","Juan","Jose"],
    "edad" : [20,23,25],
    "dni" : ["48014003P","48014003Q","48014003R"]
}

df= pd.DataFrame(personas)


## metodo 1 : Sintaxis Basica
df ["profesiones"] = ["ingeniero","maestro", "bombero"]

## metodo 2 : Assing
df= df.assign(sueldo=[20000,30e3,40e3])

## metodo 3 : insert modifica directo el objeto no se necesita reasignar como los dos anteriores
df.insert(3,"email",["dimas@mail.com","juan@email.com","jose@email.com"])
df
```

```python
# Como obtener las filas de un DataFrame en python
df
#############metodo 1 : sintaxys basica

condicion = (df ["nombre"]=="Dimas")

      ## Ver donde se cumple
print (condicion)

df [condicion]

#############metodo 2 : isin
     ### Dos formas
#condicion = df.email.isin(["dimas@mail.com","dsfsdf","jose@email.com"])
condicion = df.email.isin(["dimas@mail.com"]) | df.nombre.isin(["Jose"]) 
df [condicion] 


#############metodo 3 : iloc
    ## Por rango cualquiera de las dos formas
df.iloc[[0,1], :]
df.iloc[[2], :]

#############metodo 4 : set_index and loc
nombres = df.set_index("nombre")
print (nombres)
   
   ### primero selecciona todo y la segunda selecciona una columna comentar para ver
nombres.loc[["Jose","Juan"]]
nombres.loc[["Jose","Juan"],["dni"]]
nombres.loc[["Jose","Juan"],["dni","email"]]

```
