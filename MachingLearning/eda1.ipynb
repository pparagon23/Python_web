```python
import pandas as pd
from sklearn.datasets import load_iris

load_iris()
```

```python
## Convierte el data set en el data frame , return lo pasa a dos dataframe X = LAbel Y = Clases
X,Y = load_iris(as_frame=True, return_X_y=True)

## agregar columna en pandas
X["labels"]=Y
X.head(10)
```

```python
especies= list(load_iris()["target_names"])
especies
```

```python
## se comenta por que si se vuelve a ejecutar la columna label ya no es numerico
X["labels"]=X["labels"].apply(lambda Label: especies[Label])
X.head(10)
```

```python
## Renemobrar columnas
X.rename(
         {
          X.columns[0]: "sepal_length",
          X.columns[1]: "sepal_width",
          X.columns[2]: "petal_length",
          X.columns[3]: "petal_length",
          },
         axis="columns",
         inplace=True
         )
## Escribe en un carchivo csv
#X.to_csv("iris.csv")
X.head(10)
```

```python
import matplotlib.pyplot as plt

## fig contiene todo el grafico y ax contiene los ejes que para este caso son 4
fig,ax = plt.subplots(
    figsize=(7,7), nrows=2, ncols=2
)
fig.suptitle("Graficos de cada caracteristica")
current_ax = 0
colors=["red","green","blue","orange"]


## feature_nmae= nombre de la columna, feature_arrays= todos los elementos de la columna
for feature_name, feature_arrays in X.items():
    if feature_name == "labels":
        break
    
   ## pasar a una flla 
    ax = fig.axes[current_ax]
    
    if current_ax == 0:
        ax.plot(
          X["sepal_width"],
          label="sepal_width",
          color="purple",
          marker="x",
          linewidth=0.3,    
        )
        
        
    ax.plot(
        feature_arrays,
        label=feature_name,
        color=colors[current_ax],
        marker="o",
        linewidth=1,
    )
    ax.grid()
    ax.set_title(feature_name)
    ax.set_ylabel(feature_name)
    ax.set_xlabel("Indice de la muestra")
    ax.legend()
    ax.axis('auto')
    current_ax += 1
## ajusta la presentacion  conforme a la vista   
fig.tight_layout()

```
